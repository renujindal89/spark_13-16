# Benefits of Parquet

```python
1.Parquet stores data column-wise, not row-wise.

👉 Spark reads only required columns, not the full table.
eg.SELECT name, salary FROM employees;
✔ Reads only name and salary columns
❌ Skips other columns

2.Smaller File Size (Compression)
👉 70–90% size reduction compared to CSV/JSON
3.Better Query Performance
👉 Queries run much faster than CSV/JSON
4.Schema Enforcement(no need to write inferschema=True)
eg.✔ salary stays numeric
❌ No accidental strings like "five thousand"
5.splittable & Parallel Processing
Files can be split
Spark reads them in parallel
👉 Better cluster utilization
```

# Repartition vs Coalesce (Even vs Uneven Distribution)

```python
Repartition vs Coalesce (Even vs Uneven Distribution)
Feature	                            repartition()	                               coalesce()
Data distribution                   Even distribution                             	Uneven distribution
Shuffle	                             ✅ Full shuffle happens	                      ❌ No shuffle (by default)
Increasing partitions	                  ✅ Yes	                                     ❌ No
Decreasing partitions                   ✅ Yes                                      ✅ Yes
Performance cost	                    Higher (shuffle is expensive)	                  Lower
Use case	                      When balanced workload is required	        When just reducing partition cheaply
```

```python
repartition() → Even Distribution
Spark reshuffles all data across the worker node
Ensures each partition has roughly the same amount of data.

df_repart = df.repartition(4)
Spark redistributes all rows evenly
📊 Result (approx):
Partition 1 → 25%
Partition 2 → 25%
Partition 3 → 25%
Partition 4 → 25%
✔️ Even distribution
❌ Expensive (shuffle)

```

```python
coalesce() → Uneven Distribution
Spark merges existing partitions without reshuffling.
Some partitions may contain much more data than others.
df_coal = df.coalesce(2)
📊 Result:
Partition 1 → 70%
Partition 2 → 30%
⚠️ Uneven distribution
✔️ Fast (no shuffle)
```

```python
When to Use What?

Use repartition()
✔️ For joins, aggregations, heavy transformations
✔️ When parallelism and balance matter

Use coalesce()
✔️ Before writing output files
✔️ When reducing partitions cheaply
```

```python
Operations That Cause Shuffle
Operation	            Shuffle?
groupBy()	             ✅ Yes
reduceByKey()	        ✅ Yes
join()	                 ✅ Yes
distinct()	             ✅ Yes
repartition()	         ✅ Yes
orderBy()	            ✅ Yes
coalesce()	          ❌ No (by default)
map()	              ❌ No
filter()	            ❌ No
```

```python
#POST – Create data
import requests

post_data = {
    "userId": 1,
    "title": "Learning PySpark API",
    "body": "This is a POST request example"
}

response = requests.post(
    "https://jsonplaceholder.typicode.com/posts",
    json=post_data
)

print(response.status_code)
print(response.json())
```

```python
Because id :101,is created by the server, not by you
When the server receives a POST request:

It creates a new record

It automatically generates a unique ID

Then it sends the created object back
```

```python
What happens here:

requests.post() → makes a POST request

URL /posts → endpoint to create new data

json=post_data:

Converts Python dict → JSON

Sends it in request body

Sets Content-Type: application/json
```

# put - update


```python
#PUT – Update the SAME data

#We update post id = 1

update_data = {
    "userId": 1,
    "title": "Updated Title",
    "body": "This data is updated using PUT"
}

response = requests.put(
    "https://jsonplaceholder.typicode.com/posts/1",  # where userid =1
    json=update_data
)

print(response.status_code)
print(response.json())
```

```python
Real-world APIs (Critical concept)

In real production APIs:

id → ❌ never changeable

userId → ❌ often NOT changeable (ownership field)

title, body → ✅ changeable
```

```python
PUT vs PATCH (Interview favorite)
PUT

Replaces entire object

Usually requires all fields

PATCH

Updates only specific fields

requests.patch("/posts/1", json={"title": "New title"})
```

```python
response=requests.patch("https://jsonplaceholder.typicode.com/posts/1", json={"title": "New title"})
print(response.status_code)
print(response.json())
```

```python
#DELETE – Remove data
import requests

response = requests.delete(
    "https://jsonplaceholder.typicode.com/posts/1"  # delete post with id = 1
)

print(response.status_code)
print(response.text)
```

```python
200 OK → deleted  and get successfully (with response)
201 Created – Resource successfully created (e.g., after POST).
204 No Content → deleted successfully (no response body)
404 Not Found – Requested resource doesn’t exist.
503 Service Unavailable – Server temporarily unavailable (e.g., maintenance).
101 informational error
```
